```python
import numpy as np
```

# Ejercicio 1

```python
A = np.array([[5, 12, 3], [3, 5, 5]])
B = np.array([[235], [115], [105]])

C = A.dot(B)
print(f"la matriz C es: \n{C}\n")
```

# Ejercicio 2

```python
E = np.array([[2, 3], [5, 1]])
F = np.array([[3, 5, 7], [2, 8, 5]])

G = E.dot(F)
print(f"la matriz G es: \n{G}\n")
```

```python

E = np.array([[66, 45, 68], [70, 54, 56], [70, 49, 64]])
F = np.array([[480, 470], [580, 640], [420, 370]])

# E(3X3) * F(3X2) = G(3X2)

G = E.dot(F)
print(f"la matriz G es: \n{G}\n")
```
